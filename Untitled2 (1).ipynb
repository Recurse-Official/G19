```python
!pip install gradio
import os
import cv2
import gradio as gr
from pymongo import MongoClient
from googleapiclient.discovery import build
from googleapiclient.http import MediaFileUpload
from google.oauth2 import service_account
import insightface
from insightface.app import FaceAnalysis

# MongoDB connection setup (replace with your credentials)
MONGO_URI = "mongodb+srv://vijju070806:vijju070806@vijju0.9uta5.mongodb.net/?retryWrites=true&w=majority&appName=Vijju0"
client = MongoClient(MONGO_URI)
db = client["face_database"]
collection = db["face_embeddings"]

# Path to the credentials file for Service Account authentication
SERVICE_ACCOUNT_FILE = '/content/drive/MyDrive/wedding-hackathon-b0a242612b20.json'  # Update this path

# Scopes for accessing Google Drive
SCOPES = ['https://www.googleapis.com/auth/drive.file']

def get_google_drive_service():
    """Authenticate and return Google Drive service using a service account."""
    try:
        creds = service_account.Credentials.from_service_account_file(SERVICE_ACCOUNT_FILE, scopes=SCOPES)
        service = build('drive', 'v3', credentials=creds)
        return service
    except Exception as e:
        print(f"Error connecting to Google Drive API: {e}")
        return None

def upload_to_drive(file_path, folder_id):
    """Upload file to Google Drive."""
    service = get_google_drive_service()
    if not service:
        return None  # If service is None, return early

    try:
        # Create file metadata including the folder ID
        file_metadata = {'name': os.path.basename(file_path), 'parents': [folder_id]}
        media = MediaFileUpload(file_path, mimetype='image/jpeg')  # Assuming JPEG, adjust for other types

        # Upload the file
        file = service.files().create(body=file_metadata, media_body=media, fields='id').execute()
        print(f"Image uploaded successfully! File ID: {file.get('id')}")
        return file.get('id')
    except Exception as e:
        print(f"Error uploading to Google Drive: {e}")
        return None

# Initialize ArcFace model
app = FaceAnalysis()
app.prepare(ctx_id=0, det_size=(640, 640))

def save_face_embedding(image_path):
    """Save face embedding into MongoDB."""
    try:
        # Load image using OpenCV
        img = cv2.imread(image_path)
        if img is None:
            print(f"Error: Unable to read the image at {image_path}. Check the file path and format.")
            return

        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB format

        # Perform face detection with ArcFace
        faces = app.get(img)
        if len(faces) == 0:
            print("No faces detected in the image.")
            return

        # Save embeddings to MongoDB
        for face in faces:
            embedding_data = {
                "embedding": face.embedding.tolist(),  # Convert numpy array to list for MongoDB
                "image_path": image_path
            }
            print(embedding_data)
            # Check if an embedding for this image already exists
            existing_doc = collection.find_one({"image_path": image_path})
            if existing_doc:
                # If the document exists, update it with the new embedding
                result = collection.update_one({"_id": existing_doc["_id"]}, {"$set": embedding_data})
                if result.modified_count > 0:
                    print(f"Updated embedding for image: {image_path}")
                else:
                    print(f"No changes made for image: {image_path}")
            else:
                # Otherwise, insert the new embedding document
                collection.insert_one(embedding_data)
                print(f"Saved new embedding for image: {image_path}")
    except Exception as e:
        print(f"Error saving embedding for image {image_path}: {e}")

def save_and_upload_images(image_paths, drive_folder_id):
    """Save face embeddings and upload the images to Google Drive."""
    for image_path in image_paths:
        # First, save the embedding
        save_face_embedding(image_path)

        # Upload image to Google Drive
        file_id = upload_to_drive(image_path, drive_folder_id)
        if file_id:
            print(f"Image uploaded to Google Drive successfully! File ID: {file_id}")
        else:
            print(f"Failed to upload image to Google Drive.")

# Gradio interface function
def process_images(image_files):
    """Process uploaded images, save embeddings, and upload to Google Drive."""
    # Convert file paths from the uploaded files
    image_paths = [file.name for file in image_files]

    # Set your Google Drive folder ID here
    drive_folder_id = "1-hrbBL-gcPzKpN2zlB-fSO1OOXb2MMay"

    # Save and upload images
    save_and_upload_images(image_paths, drive_folder_id)

    return "Images processed successfully!"

# Gradio interface setup
# Gradio interface setup
iface = gr.Interface(
    fn=process_images,
    inputs=gr.File(file_count="multiple", type="filepath"),
    outputs="text",
    live=True
)

# Launch the interface
iface.launch()
```
